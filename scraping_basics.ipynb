```python
from bs4 import BeautifulSoup
import requests
```

```python
with open('dataset/wiki_test.htm') as html_file:
    soup = BeautifulSoup(html_file, 'lxml')
```

```python
print(soup.prettify())
```

```python
print(soup.title.text)
```

```python
soup.find('div', class_='mw-indicators')
## class is a special word
```

```python
for s in soup.find_all('span', class_="toctext"):
    print(s.text)
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 
## Keys special keys code for web control 
```

```python
driver = webdriver.Chrome(executable_path='/Users/violet/Downloads/chromedriver')
## chrome needs chromediver for execution 
driver.get('http://www.python.org')
## webdriver will wait until the page is fulled loaded, Ajax may confuse the webdriver 
```

```python
assert 'Python' in driver.title 
```

```python
elem = driver.find_element_by_name('q')
## find_element_by_id('id-search-field')
elem.clear()
## clear the text entry element 
elem.send_keys('selenium')
elem.send_keys(Keys.RETURN)
## stimulates typing into the element
## Keys.RETURN: capitalized 
```

```python
assert 'No results found' not in driver.page_source 
## driver.page_source: current page 
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC 
```

```python
driver = webdriver.Chrome(executable_path='/Users/violet/Downloads/chromedriver')
driver.get("http://somedomain/url_that_delays_loading")
try: 
    elem = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((BY.ID,"myDynamicElement"))
    )
except: 
    print('Error')
finally: 
    driver.close()
```
